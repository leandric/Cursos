# Imports

Aqui importamos as bibliotecas que instalamos para usar no código

```python
import polars as pl
import fsspec
import io
```

Esse trecho está criando uma conexão com um sistema de arquivos remoto via SFTP, usando o fsspec como camada de abstração.

“Me dê um objeto que saiba acessar arquivos via SFTP nesse servidor.”

```python
fs = fsspec.filesystem("sftp", 
                       host="localhost", 
                       port=2222, 
                       username="teste", 
                       password="123456")
```

Esse trecho faz a leitura de um arquivo remoto (via SFTP) e converte os bytes em texto legível.

```python
with fs.open("/upload/microdados_enem_2024/DADOS/RESULTADOS_2024.csv", "rb") as f:
    content = f.read()

text = content.decode("latin-1")
```

Lê o conteúdo de texto do CSV (já decodificado) a partir de um buffer em memória (`StringIO`) e o transforma em um DataFrame do Polars usando `;` como separador.


```python
df_resultado = pl.read_csv(io.StringIO(text), separator=";")
```

executando a variavel que contém o dataframe o jupyter print uma pequena parte da tabela, notem que no canto superior esquerdo temos o shape (quantidade de linhas e colunas) da tabela

```python
df_resultado
```

# Como fazer Filtros e Selecionar colunas em um daframe polars

Para selecionar colunas usamos o metodo .select() e passamos no parametro o nome das colunas conforme o exemplo

```python
df_resultado.select(
    'NU_SEQUENCIAL',
    'NU_ANO',
    'CO_ESCOLA',
    'CO_MUNICIPIO_ESC',
    'NO_MUNICIPIO_ESC',
    'CO_UF_ESC'
)
```

também temos como selecionar usando o método .col() do polars, a diferença pro anterior é que usando o método .col() podemos usar uma serie de outros métodos, ou seja temos mais opções de manipulação

```python
df_resultado.select(
    pl.col("NU_SEQUENCIAL").alias("chave primária"),
    pl.col("NU_ANO").alias("Ano"),
    pl.col("CO_ESCOLA").alias("codigo da escola"),
    pl.col("CO_MUNICIPIO_ESC"),
    pl.col("NO_MUNICIPIO_ESC"),
    pl.col("CO_UF_ESC"),
)
```
